```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

# Distribution and Categorical plots


```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
#kde=True for trendline
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#stats.norm for normal distribution with rugplot
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) 
#by default takes mean
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#Show quartiles,
#points outside whiskers are outliers
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex',split=False)
#split=True for distribution of hue in one plot itself
```

```python
sns.stripplot(x='day', y='total_bill', data=tips,jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
#comb of violin and stripplot
```

# Matrix Plots

```python
tc=tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot_table(index='month',columns='year', values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='magma')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white', linewidth=1)
```

```python
#hierarchical
sns.clustermap(fp)
#standard_scale=1 for rescaling to 0-1
```
