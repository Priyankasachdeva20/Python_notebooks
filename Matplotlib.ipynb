```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig=plt.figure(figsize=(8,4))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X squared")
ax.plot(x,x**3, label="X cubed")
ax.legend(loc=0)
```
