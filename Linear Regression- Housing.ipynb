```python
X=df[["GarageArea","TotalBsmtSF","MSSubClass","OverallQual","YearBuilt","MoSold", "YrSold","OpenPorchSF", "WoodDeckSF","GrLivArea",
      "BedroomAbvGr","FullBath","HalfBath","TotRmsAbvGrd","BsmtFullBath","BsmtHalfBath"]]
Y=df[["SalePrice"]]
```

```python
#!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(X_train,Y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
cdf=pd.DataFrame(lm.coef_.transpose(), X_train.columns ,columns=['Coeff'])
```

```python
cdf
```

# Predictions


```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(Y_test,predictions)
```

```python
sns.distplot((Y_test-predictions))
#Histogram of residuals - normally distributed means good choice of model/data
```

# Evaluations

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(Y_test,predictions))
print(metrics.mean_squared_error(Y_test,predictions))
print(np.sqrt(metrics.mean_squared_error(Y_test,predictions)))
```
